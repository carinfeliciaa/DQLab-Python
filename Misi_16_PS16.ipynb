# Correlation

In this DataFrame, the data types are nominal, ordinal, and ratio. Correlation test is done to find out how strong the relationship is between two variables.

```python
#ordinal encoding
df['diet_quality_encoded'] = df['diet_quality'].map({'Fair': 1, 'Good': 2, 'Poor': 3})
df['internet_encoded'] = df['internet_quality'].map({'Poor': 1, 'Average': 2, 'Good': 3})
df['parental_encoded'] = df['parental_education_level'].map({'Bachelor': 1, 'High School': 2, 'Master': 3, 'Master': 4})
```

mental_health_rating and exercise_frequency are not converted since they are already in the form of number

```python
ordinal_cols = [
    'diet_quality_encoded',
    'exercise_frequency',
    'parental_encoded',
    'internet_encoded',
    'mental_health_rating',
    'exam_score']

#spearman correlation matrix
sns.heatmap(df[ordinal_cols].corr(method='spearman'), annot=True, cmap='YlGnBu')
```

```python
#dataset for ratio
ratio_df = df[['study_hours_per_day', 'sleep_hours', 'social_media_hours',
                'netflix_hours', 'attendance_percentage', 'exam_score']]
```

```python
#check the correlations
sns.heatmap(ratio_df.corr(method='spearman'), annot=True, cmap='YlGnBu')
```

# ANOVA

Are the mean exam scores different between students who have part-time jobs and students who don not have part-time jobs?

```python
import scipy.stats as stats

# ANOVA between 'part_time_job' categories and exam_score
anova_result = stats.f_oneway(
    df[df['part_time_job'] == 'Yes']['exam_score'],
    df[df['part_time_job'] == 'No']['exam_score']
)

print("ANOVA F-statistic:", anova_result.statistic)
print("p-value:", anova_result.pvalue)
```

There is no significant difference in exam scores between students who have part-time jobs and those who don not have. The high p-value (0.6105) shows that any difference is likely just due to chance, not because of having a part-time job.

Are the mean exam scores different between students who have extracurricular and students who do not have extracurricular activities?

```python
import scipy.stats as stats

# ANOVA between 'extracurricular_participation' categories and exam_score
anova_result = stats.f_oneway(
    df[df['extracurricular_participation'] == 'Yes']['exam_score'],
    df[df['extracurricular_participation'] == 'No']['exam_score']
)

print("ANOVA F-statistic:", anova_result.statistic)
print("p-value:", anova_result.pvalue)
```

There is no significant difference in exam scores between students who have
extracurricular. The high p-value (0.9174) shows that any difference is likely just due to chance, not because of having an extracurricular.

Are the mean exam scores different between genders?

```python
import scipy.stats as stats

# ANOVA between 'gender' categories and exam_score
anova_result = stats.f_oneway(
    df[df['gender'] == 'Male']['exam_score'],
    df[df['gender'] == 'Female']['exam_score'],
    df[df['gender'] == 'Other']['exam_score']
)

print("ANOVA F-statistic:", anova_result.statistic)
print("p-value:", anova_result.pvalue)
```

There is a significant difference in exam scores between genders. The high p-value (0.5051) might likely happen on a random chance.
